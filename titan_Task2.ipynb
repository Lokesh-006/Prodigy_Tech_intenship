# Imputing the null Age VAlues by 24 bcz its has occured and 24.

```python
titan_data['Age'].fillna(24,inplace=True)
```

```python
fig=px.box(titan_data,y='Age')
fig.show()
```

```python
z_scores = stats.zscore(titan_data['Age'])
print(z_scores)
```

```python
# Find outliers using z-score
z_scores = stats.zscore(titan_data['Age'])
outlier_indices = abs(z_scores) > 3  # Get indices of outliers
outlier_values = titan_data.loc[outlier_indices, 'Age']  # Get values of outliers

# Replace outliers with the mean
titan_data.loc[outlier_indices, 'Age'] = titan_data['Age'].mean()  # Replace values at outlier indices

# Verify that outliers are replaced
fig = px.box(titan_data, y='Age')
fig.show()
```

```python
fig=px.pie(titan_data,names='Cabin')
fig.show()
```
